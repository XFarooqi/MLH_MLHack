```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Read the csv file, and check its top 10 rows
pokemon = pd.read_csv('pokemon.csv')
print(pokemon.shape)
pokemon.head(10)
```

# Barchart

```python
# Count the Pokemon on the basis of thier types
sb.countplot(data=pokemon, y='type_1');
```

# Histogram

```python
# Histogram on basis of Speed
plt.hist(data = pokemon, x = 'speed')
```
